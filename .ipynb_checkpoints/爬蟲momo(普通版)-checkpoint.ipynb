```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv
import re

# 設定 Chrome 瀏覽器選項
chrome_options = Options()
chrome_options.add_argument("--headless")  # 無頭模式
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument('User-Agent="Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.70 Safari/537.36"')

# 啟動 ChromeDriver
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

# 訪問目標網頁
url = 'https://www.momoshop.com.tw/goods/GoodsDetail.jsp?i_code=10533160&Area=search&mdiv=403&oid=1_7&cid=index&kw=%E5%A4%A7%E5%90%8C%E9%9B%BB%E9%8D%8B'
driver.get(url)

time.sleep(5)  # 確保頁面完全載入

# 點擊商品評價按鈕
try:
    review_button = WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//li[contains(@class, 'goodsCommendLi')]//span"))
    )
    driver.execute_script("arguments[0].click();", review_button)  # 使用 JavaScript 點擊
    time.sleep(3)  # 等待評論加載
except Exception as e:
    print("無法找到評論按鈕: ", e)
    driver.quit()
    exit()

# 獲取總頁數
try:
    page_number_element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, 'div.pageArea span'))
    )
    print(f"頁數資訊: {page_number_element.text}")  # Debugging
    total_pages_match = re.search(r'(\d+)/(\d+)', page_number_element.text)
    total_pages = int(total_pages_match.group(2)) if total_pages_match else 48
    print(total_pages)
except Exception as e:
    print("無法獲取總頁數: ", e)
    total_pages = 48  # 預設為 48 頁

# 開始抓取評論與評分
comments_data = []
ratings_data = []

def scrape_page():
    WebDriverWait(driver, 10).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'div.reviewCard'))
    )
    time.sleep(3)  # 避免評論還未完全加載

    # 爬取評論
    comments = driver.find_elements(By.CSS_SELECTOR, 'div.reviewCardInner div.CommentContainer:not(.ReplyContainer .CommentContainer) p.Comment')
    for comment in comments:
        comments_data.append(comment.text)

    # 爬取評分
    rating_stars = driver.find_elements(By.CSS_SELECTOR, 'div.RatingStarGroup')
    for star in rating_stars:
        score = star.get_attribute('score')
        ratings_data.append(score)

# 抓取第一頁評論
scrape_page()

# 抓取剩餘頁面評論
for page in range(2, total_pages + 1):
    try:
        print(f"正在爬取第 {page} 頁...")

        # **修正 XPath 來選擇分頁按鈕**
        next_page_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, f"//dd[@pageidx='{page}']/a"))
        )

        # **確保按鈕可以被點擊**
        driver.execute_script("arguments[0].click();", next_page_button)

        # **等待新頁面載入**
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'div.reviewCard'))
        )
        time.sleep(3)  # 避免評論還未完全加載

        # 爬取新頁面的評論
        scrape_page()

    except Exception as e:
        print(f"無法點擊第 {page} 頁: ", e)
        break



# 關閉瀏覽器
driver.quit()
```

```python
comments_data = [[comment] for comment in comments_data]
ratings_data = [[rating] for rating in ratings_data]
```
